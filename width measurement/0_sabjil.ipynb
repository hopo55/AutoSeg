```python
def get_rotation_angle(image):
    # Take the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Fourier Transform
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)

    # Find the magnitude spectrum, log transform it, and normalize it
    magnitude_spectrum = np.log(np.abs(fshift))

    # Find the orientation of the main frequency component
    orientation = ndimage.measurements.center_of_mass(magnitude_spectrum)

    # This will give us the angle in radians in range [-pi, pi]
    angle_radians = np.arctan2(orientation[0], orientation[1])

    # Convert to degrees and normalize to range [0, 180]
    angle_degrees = np.degrees(angle_radians) % 180

    # Return the dominant angle of rotation
    return angle_degrees
```
